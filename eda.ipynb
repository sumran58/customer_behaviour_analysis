## treating null values

### size

```python
df['Size'].unique()
```

```python
df.groupby('Category')['Size'].unique()
```

```python
#we can see in the electronics category also we have the sizes like m,l,xl but logically this is not correct so we need to replace it 
df.loc[df['Category']=='Electronics','Size']="Not Applicable"
```

```python
df.groupby('Category')['Size'].unique()
```

```python
#the Acessories column also has the sizes and generally this accecories also dont have any size so we will replace it with Free Size
df.loc[df['Category']=='Accessories','Size']="Free Size"
```

```python
df.groupby('Category')['Size'].unique()
```

```python
#footwear column also contains teh sizes but logically it cant be soi w will repolace it with not available
df.loc[df['Category']=='Footwear','Size']="Not Available"
```

```python
df.groupby('Category')['Size'].unique()
```

```python
df.loc[(df['Category']=='Clothing')&(df['Size'].isnull()),'Size']=df[df['Category']=='Clothing']['Size'].mode()[0]
#jaha pe category clothing rahegi and siz column me values jaha null rahegi bas wahi  we ahve to replace the nan values with the mode value and [0] isliye liye because there can be tie of the mode value therefore we are taking the value from the top position 
```

```python
df.isnull().sum()
```

```python
df.groupby("Item Purchased")['Review Rating'].mean()
```

```python
#replacing the nan values in the review rating based on the means of the item purchased
df['Review Rating']=df['Review Rating'].fillna(df.groupby('Item Purchased')['Review Rating'].transform('mean'))
```

```python
df['Review Rating']=df['Review Rating'].round(2)
```

```python
df.isnull().sum()
```

```python
df['Previous Purchases']=df['Previous Purchases'].fillna(0)
```

```python
df.isnull().sum()
```

```python
df['Purchase Amount (USD)']=df['Purchase Amount (USD)'].fillna(df.groupby('Item Purchased')['Purchase Amount (USD)'].transform('mean'))
```
